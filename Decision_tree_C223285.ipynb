<h2><b>Decision Tree</b></h2>

Step 1: Import Libraraies

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

Step 2: Load the Dataset

```python
data=pd.read_csv("/content/sample_data/spam_ham_dataset.csv")

data.head(5)
```

```python
#Shape of the data
print("Number of rows",data.shape[0])
print("Number of columns",data.shape[1])
```

Step 3: Prepare Features X and Target Y

```python
X=data[['text']]
Y=data['label_num']

print("Features: ", X.head())

print("\nTarget: ", Y.head())
```

Step 4: Split Data into Training and testing sets

```python
#70% for training and 30% for testing
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=50)

#The random_state parameter in train_test_split is used to ensure reproducibility. When you set a specific integer value for random_state, the dataset split will be the same every time you run your code. If you don't set it, or set it to None, the split will be different each time, which can make it hard to compare model performance across different runs.
print("Training set size: ", len(X_train))
print("Testing set size: ", len(X_test))
```

Step 5: Create and Train the Decision Tree Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training data, then transform the test data
X_train_vectorized = vectorizer.fit_transform(X_train['text'])
X_test_vectorized = vectorizer.transform(X_test['text'])

#create model
model=DecisionTreeClassifier(random_state=50)

#train model
model.fit(X_train_vectorized,Y_train)
```

Step 6: Predictions

```python
Y_predict=model.predict(X_test_vectorized)

print("Actual Values: ", list(Y_test))
print("Predicted Values: ", list(Y_predict))
```

Step 7: Evaluate the model

```python
#accuracy calculation
accuracy=accuracy_score(Y_test,Y_predict)
print("Accuracy: ", round(accuracy*100,2),"%")

#classification report
report=classification_report(Y_test,Y_predict)
print("\nClassification Report: \n", report)

#confusion matrix
confusion_mat=confusion_matrix(Y_test,Y_predict)
print("\nConfusion Matrix: \n", confusion_mat)
```

Step 8: Important feature for decision

```python
#get feature names
feature_names = ['label', 'text', 'label_num']

#get feature importances
importances = model.feature_importances_

#Display feature importance
print("Feature Importance:\n")
for name, score in zip(feature_names, importances):
   print(name, ":", round(score * 100, 2), "%")
```

Step 9: Visualize the decision tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot the decision tree
plt.figure(figsize=(15, 10))

#plot the tree
plot_tree(model,
          feature_names=vectorizer.get_feature_names_out(),
          class_names=['ham', 'spam'],
         filled=True,
         max_depth=2 # Added max_depth to make the plot more readable for a high-dimensional feature space.
         )

# Show the plot
plt.title("\nDecision Tree\n")
plt.show()
```

Step 10: Control Tree Depth to Prevent Overfitting

```python
#Create a simple tree with max dept=3
simple_model=DecisionTreeClassifier(max_depth=3,random_state=50)

#train model
simple_model.fit(X_train_vectorized,Y_train)

#Predict
Y_predict=simple_model.predict(X_test_vectorized)

#accuracy calculation
accuracy_simple=accuracy_score(Y_test,Y_predict)
print("Accuracy: ", round(accuracy*100,2),"%")
```
