```python
#Exercise 1: Create a function to calculate Mean Squared Error (MSE)
# TODO: Implement MSE function
def calculate_mse(y_true, y_pred):
    """
    Calculate Mean Squared Error

    Parameters:
    -----------
    y_true : list
        True values
    y_pred : list
        Predicted values

    Returns:
    --------
    float
        Mean squared error
    """
    n=len(y_true)
    mse=0
    for i in range(n):
        mse+=(y_true[i]-y_pred[i])**2
    mse=mse/n
    return mse
    pass

# Test your function
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
# Expected MSE â‰ˆ 0.375
mse = calculate_mse(y_true, y_pred)
print(f"MSE: {mse}")
```

```python
#Exercise 2: Create a batch generator function
def create_batches(data, batch_size):
    """
    Create batches from data

    Parameters:
    -----------
    data : list
        Input data
    batch_size : int
        Size of each batch

    Returns:
    --------
    list
        List of batches
    """
    batches = []
    for i in range(0, len(data), batch_size):
        batches.append(data[i:i + batch_size])
    return batches

# Test your function
data = list(range(1, 11))  # [1, 2, 3, ..., 10]
batches = create_batches(data, batch_size=3)
print(f"Batches: {batches}")
```

```python
# Exercise 3: Implement a confusion matrix calculator
def confusion_matrix(y_true, y_pred):
    """
    Calculate confusion matrix for binary classification

    Parameters:
    -----------
    y_true : list
        True labels (0 or 1)
    y_pred : list
        Predicted labels (0 or 1)

    Returns:
    --------
    dict
        Dictionary with TP, TN, FP, FN
    """
    tp = 0
    tn = 0
    fp = 0
    fn = 0

    for i in range(len(y_true)):
        if y_true[i] == 1 and y_pred[i] == 1:
            tp += 1
        elif y_true[i] == 0 and y_pred[i] == 0:
            tn += 1
        elif y_true[i] == 0 and y_pred[i] == 1:
            fp += 1
        elif y_true[i] == 1 and y_pred[i] == 0:
            fn += 1

    return {'TP': tp, 'TN': tn, 'FP': fp, 'FN': fn}

# Test your function
y_true = [1, 0, 1, 1, 0, 1, 0, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1, 0]
cm = confusion_matrix(y_true, y_pred)
print(f"Confusion Matrix: {cm}")
# Expected: {'TP': 3, 'TN': 3, 'FP': 1, 'FN': 1}
```

```python
#Exercise 2: Pandas Data Filtering
# TODO: Using the students_df dataset, complete the following:
import pandas as pd
import numpy as np
n_students = 30
students_data = {
    'StudentID': [f'S{i:03d}' for i in range(1, n_students + 1)],
    'Name': [f'Student_{i}' for i in range(1, n_students + 1)],
    'Age': np.random.randint(18, 25, n_students),
    'Gender': np.random.choice(['Male', 'Female'], n_students),
    'Math': np.random.randint(60, 100, n_students),
    'Physics': np.random.randint(55, 100, n_students),
    'Chemistry': np.random.randint(50, 100, n_students),
    'English': np.random.randint(65, 100, n_students),
    'StudyHours': np.random.randint(1, 10, n_students),
    'Attendance': np.random.randint(70, 100, n_students)
}
students_df = pd.DataFrame(students_data)
print("Original DataFrame:\n", students_df.head())
print()

# 1. Find all students who scored above 85 in Math AND Physics
high_scorers = students_df[(students_df['Math'] > 85) & (students_df['Physics'] > 85)]
print("1. Students with Math > 85 and Physics > 85:\n", high_scorers.head())
print()

# 2. Calculate the average score for students who study more than 5 hours
high_study_hours_students = students_df[students_df['StudyHours'] > 5]
average_score_high_study = high_study_hours_students[['Math', 'Physics', 'Chemistry', 'English']].mean().mean()
print(f"2. Average score for students who study more than 5 hours: {average_score_high_study:.2f}")
print()

# Calculate overall average for grading purposes
students_df['Average'] = students_df[['Math', 'Physics', 'Chemistry', 'English']].mean(axis=1)

# 3. Find students with Grade 'A+' or 'A' and attendance > 90%
# Assuming 'A+' or 'A' means an average score >= 85 (as per 'Excellent' in task 4)
grade_a_students = students_df[(students_df['Average'] >= 85) & (students_df['Attendance'] > 90)]
print("3. Students with Grade 'A' (Average >= 85) and Attendance > 90%:\n", grade_a_students.head())
print()

# 4. Create a new column 'Performance' with values:
#    'Excellent' if Average >= 85
#    'Good' if Average >= 75
#    'Average' if Average >= 65
#    'Below Average' otherwise
def assign_performance(row):
    if row['Average'] >= 85:
        return 'Excellent'
    elif row['Average'] >= 75:
        return 'Good'
    elif row['Average'] >= 65:
        return 'Average'
    else:
        return 'Below Average'

students_df['Performance'] = students_df.apply(assign_performance, axis=1)
print("4. DataFrame with 'Performance' column (first 5 rows):\n", students_df[['Name', 'Average', 'Performance']].head())
```

```python
# Practice dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(100)
practice_df = pd.DataFrame({
    'Age': np.random.randint(20, 60, 100),
    'Salary': np.random.randint(30000, 120000, 100),
    'Experience': np.random.randint(0, 20, 100),
    'Department': np.random.choice(['IT', 'Sales', 'HR'], 100)
})

df=practice_df.copy()
print(practice_df.head())

# TODO: Your tasks
# 1. Create a scatter plot: Experience vs Salary
# 2. Create a box plot: Salary by Department
# 3. Create a histogram: Age distribution
# 4. Create a correlation heatmap for numerical columns

# Your code here:
#1
plt.figure(figsize=(8, 5))
sns.regplot(data=df, x='Experience', y='Salary', scatter_kws={'alpha':0.6})
plt.title('Experience vs Salary', fontsize=14, fontweight='bold')
plt.show()


#2
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Department', y='Salary', palette='Set1')
plt.title('Salary by Department', fontsize=14, fontweight='bold')
plt.show()

#3
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=15, color='blue', alpha=0.7, edgecolor='black')
plt.title('Age distribution', fontsize=14, fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(df['Age'].mean(), color='red', linestyle='--', linewidth=2, label=f"Mean: {df['Age'].mean():.1f}")
plt.legend()
plt.show()

#4
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Numerical Columns', fontsize=14, fontweight='bold')
plt.show()
```
