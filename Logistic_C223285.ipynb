<h1><b>Logistic Regression</b></h1>

<h5>Step 1: Import Libraries</h5>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

<h5>Step 2: Load the data</h5>

```python
#Load the dataset
data=pd.read_csv('/content/sample_data/spam_ham_dataset.csv')

#Display First 10 rows
data.head(10)
```

```python
print(f'Number of Rows: {data.shape[0]}')
print(f'Number of Columns: {data.shape[1]}')
```

```python
data.info()
```

<h5>Step 3: Prepare Features X and Target Y</h5>

```python
#X= input features
#Y=Target

X=data['text']
Y=data['label']

print("Features (X):")
print(X.head())
print("\nTarget (Y):")
print(Y.head())
```

Step 4: Split Data into Training and Testing sets

```python
#60% for training ,40% for testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=50)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}\n')
print(f'Training sets:{len(X_train)}')
print(f'Testing sets:{len(X_test)}\n')

print(f'Y_train shape: {Y_train.shape}')
print(f'Y_test shape: {Y_test.shape}')
```

Step 5: Create and train the model

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Transform text data to numerical features
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

#model
model=LogisticRegression()

#train the model with vectorized data
model.fit(X_train_vectorized, Y_train)

print('Model trained Successfully')
```

Step 6: Predictions

```python
#predict test
Y_pred=model.predict(X_test_vectorized)

print(f'Actual values: {Y_test}')
print(f'Predicted Values: {Y_pred}')

print(f'\nActual values: ',list(Y_test))
print(f'Predicted Values: ', list(Y_pred))
```

<h5>Step 7: Evaluate the Model</h5>

```python
#calculate accuracy
accuracy=accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {accuracy*100} %')
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, Y_pred)
```

```python
#detailed Classification report
from sklearn.metrics import classification_report
print("classification_report:")
print(classification_report(Y_test, Y_pred))
```

<h5>Step 8: Predict for a new email</h5>

```python
new_mail = [
    "Subject: Claim your prize now! You have won a lottery. Click here to get your reward.",
    "Subject: Meeting reminder for tomorrow at 10 AM. Please review the attached document."
]

prediction=model.predict(vectorizer.transform(new_mail))
print(prediction)
#print(new_mail)
```
