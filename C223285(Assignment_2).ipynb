<h2><i><u><b>Linear Regression</b></u></i></h2>

## **Step 1: Import Libraries**

```python
import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
from sklearn.linear_model import LinearRegression

print("Libraries imported successfully!")
```

# **Step 2: Load the Data**

```python
#Load data from CSV file
dataset=pd.read_csv('/content/sample_data/House Price Prediction.csv')

#Display the data
print("Dataset:")
print(dataset)
```

# **Step 3: Visualization**

```python
#Scatter plot
plt.figure(figsize=(12,8))
plt.scatter(dataset['House Size (sq ft)'],dataset['Price($1000s)'],color='green',s=200)
plt.xlabel('House Size',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.title('House Size vs Price',fontsize=20,fontweight='bold',color='blue')
plt.grid(True,alpha=0.4)  #alpha indication
plt.show()
```

# **Step 4: Prepare Data for Trainig**



*   X= House Size
*   Y= Price



```python
#X=Input feature(House Size)
X=dataset['House Size (sq ft)']

#Y=Target variable(Price)
Y=dataset['Price($1000s)']

# print("Input Feature (X):")
# print(X)
# print("\nTarget Variable (Y):")
# print(Y)
print("Data Prepared for training")
```

# **Step 5: Create and Train the Model**

```python
#Creating model
model=LinearRegression()

#Train the model with our data
model.fit(X.values.reshape(-1, 1), Y) #****

print("Model trained successfully!")
```

# **Step 6: Understand the model equation**

**Y = b₀ + b₁X**

```python
#slope b₁ and intercept b₀
slope= model.coef_[0]
intercept=model.intercept_

print("Model Equation:")
print(f"Price = {slope:.3f} * House Size + {intercept:.2f}")
print(f"\nIntercept: {intercept:.2f}")
print(f"\nSlope: {slope:.3f}")
```

# **Step 7: Visualize the model**

Draw the best fit line

```python
#Make predictions for all our data points
Y_prediction= model.predict(X.values.reshape(-1, 1))   #****why values.reshape is coming?

#***what the dot and line indicating? its not showing in scatter diagram****

#Create a beautiful plot
plt.figure(figsize=(12,8))

#plot the actual data
plt.scatter(dataset['House Size (sq ft)'], dataset['Price($1000s)'], color='green', s=200, label='Actual Data', zorder=4)

#plot the prediction line
plt.plot(dataset['House Size (sq ft)'], Y_prediction,color='red', linewidth=3, label='Best Fit Line', linestyle='--',zorder=2) #use of zorder

plt.xlabel('House Size',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.title('House Size vs Price',fontsize=20,fontweight='bold',color='blue')
plt.grid(True,alpha=0.4)
plt.show()
```

# **Step 8: Predictions**

```python
new_house=1500
predicted_price=model.predict([[new_house]])[0]

print(f"Prediction: \n")
print(f" a new house size is {new_house},")
print(f"the predicted price will be: {predicted_price: .3f}")
```

```python
new_house2=2500
predicted_price=model.predict([[new_house2]])[0]

print(f"Prediction: \n")
print(f" a new house size is {new_house2},")
print(f"the predicted price will be: {predicted_price: .3f}")
```

# Multiple Prediction

```python
test_size=[1500,2500]

print("Multiple Prediction")

for size in test_size:
  predicted_price=model.predict([[size]])[0]
  print(f"\n a new house size is {size},")
  print(f"the predicted price will be: {predicted_price: .3f}")
```

# **Step 10: Performance Evaluation**

```python
#R² score (accuracy measure)
#R² score ranges from 0 to 1, where 1 = perfect prediction

r2_score=model.score(X.values.reshape(-1, 1), Y)

print("Model Performance")
print(f"R² Score: {r2_score:.3f}")
print(f"Accuracy: {r2_score*100:.3f}%")

if r2_score > 0.9:
    print("\nExcellent!")
elif r2_score > 0.7:
    print("\n Good! ")
else:
    print("\nOur model could be improved with more data or features.")
```
